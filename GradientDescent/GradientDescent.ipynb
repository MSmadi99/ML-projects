```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator
from itertools import product
```

```python
def func(x):
    return x**2 - 2*x - 3

def der_func(x):
    return 2*x - 2

def plotFunc(x0):
    x = np.linspace(-5, 7, 100)
    plt.plot(x, func(x))
    plt.plot(x0, func(x0), 'ko')
    plt.xlabel('$x$')
    plt.ylabel('$f(x)$')
    plt.title('Objective function')

def plotPath(xs, ys, x0):
    plotFunc(x0)
    plt.plot(xs, ys, linestyle='--', marker='o', color='#f12f79')
    plt.plot(xs[-1], ys[-1], 'ko')

def GradientDescentSimple(func, der_func, x0, alpha, tol=1e-5, max_iter=1000):
    xk = x0
    fk = func(xk)
    pk = -der_func(xk)

    num_iter = 0
    curve_x = [xk]
    curve_y = [fk]

    while abs(pk) > tol and num_iter < max_iter:
        xk = xk + alpha*pk
        fk = func(xk)
        pk = -der_func(xk)

        num_iter =+ 1
        curve_x.append(xk)
        curve_y.append(fk)

    if num_iter == max_iter:
        print('Gradient Descent does not converge')
    else:
        print(f'Solution found:\n y = {fk: .4f}\n x = {xk: .4f}')

    return curve_x, curve_y
```

```python
x0 = -4
xs, ys = GradientDescentSimple(func, der_func, x0, alpha=0.1)
plotPath(xs, ys, x0)
```

```python
xs, ys = GradientDescentSimple(func, der_func, x0, alpha=0.9)
plotPath(xs, ys, x0)
```

```python
xs, ys = GradientDescentSimple(func, der_func, x0, alpha=1e-4)
plotPath(xs, ys, x0)
```

```python
xs, ys = GradientDescentSimple(func, der_func, x0, alpha=0.1, max_iter=8)
plotPath(xs, ys, x0)
```
